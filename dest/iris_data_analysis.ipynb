# Iris 数据集分析

本笔记本将加载Iris数据集，预览数据，并绘制花瓣与萼片尺寸关系的散点图。

```python
# 导入必要的库
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置图表风格
# sns.set(style="whitegrid")
```

```python
# 加载Iris数据集
iris = sns.load_dataset("iris")

# 预览数据
print(iris.head())
```

```python
# 绘制散点图：花瓣长度与萼片长度的关系
plt.figure(figsize=(10, 6))
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
plt.title("花瓣长度与萼片长度的关系")
plt.xlabel("萼片长度 (cm)")
plt.ylabel("花瓣长度 (cm)")
plt.show()
```
